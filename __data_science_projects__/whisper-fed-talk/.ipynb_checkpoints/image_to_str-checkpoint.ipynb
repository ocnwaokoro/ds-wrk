```python
import base64
import re
```

```python
with open('candlestix.png','rb') as f:
    encode = str(base64.b64encode(f.read()))
```

```python
m = re.match("(.+)(b')(.+)(')",f'data:image/png;base64,{encode}')
```

```python
with open('candlestix.txt','w') as f:
    f.write(m[1] + m[3])
```
