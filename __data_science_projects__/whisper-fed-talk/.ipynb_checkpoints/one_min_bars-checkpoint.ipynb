````python
import config
import pandas as pd
from datetime import datetime
import alpaca_trade_api as tradeapi

# time in RFC-3339 format

api = tradeapi.REST(config.API_KEY,config.API_SECRET,base_url=config.API_URL)
# start 1 min before 
start = 'T'.join(str(datetime(2022, 11, 2, 14, 30, 15)).split(' ')) + 'Z'
# start 1 min after 
end = 'T'.join(str(datetime(2022, 11, 2, 15, 16, 25)).split(' ')) + 'Z'
barsets = api.get_bars(['SPY'],'1Min',start,end)
````

````python
df = pd.DataFrame([[datetime.strptime(b['t'], '%Y-%m-%dT%H:%M:%SZ'),
                    b['o'],
                    b['h'],
                    b['l'],
                    b['c'],
                    b['v'],
                    b['vw'],
                    b['n']] for b in barsets._raw])
df.columns = ['Date','Open', 'High', 'Low', 'Close', 'Volume','Avg','Num']
display(df)
````

````python
df.to_csv('spy.csv',index =False)
````

```
t	string/timestamp	Timestamp in RFC-3339 format with nanosecond precision
o	number	Open price
h	number	High price
l	number	Low price
c	number	Close price
v	int	Volume
n	int	Number of trades
vw	number	Volume-weighted average price
```
